```python
'''UPLOADING AND PRINTING DATASET'''


#upload in drive first

#Train Dataset
df_train=pd.read_csv('Train.csv')
print(df_train)


#Test Dataset
df_test=pd.read_csv('Test.csv')
print("Test")
print(df_test)

#data maniupulation to make it organise it and make it easy to read
#data belending: big data from multiple sources are mergerd into a single file
```

```python
'''# 1.MISSING VALUES HANDLING'''

#checking for missing values:(Numeric(using mean) and catagorical(using mode) missing data need to be handled differently)
#item_outlet_sales is target label, ie sales of each bigmart outlet
#Item_weight to be filled with mean
#Outlet_size to be filled with mode


df_train.isnull().sum()
df_test.isnull().sum()


df_train.describe()#gives statistcal analysis of numerical columns only


#filling Item_weight with mean
df_train['Item_Weight'].fillna(df_train['Item_Weight'].mean(),inplace=True)
df_test['Item_Weight'].fillna(df_test['Item_Weight'].mean(),inplace=True)

#filling Outlet_size using mode
print(df_train['Outlet_Size'].value_counts())#displays count of each category
df_train['Outlet_Size'].fillna(df_train['Outlet_Size'].mode()[0], inplace=True)
df_test['Outlet_Size'].fillna(df_test['Outlet_Size'].mode()[0], inplace=True)
```

```python
#2.SELECTING FEATURES DEPENDING ON GENERAL REQUIREMENTS(Dimensionality Reduction), removing irrevelant columns


df_train.drop(['Item_Identifier','Outlet_Identifier'], axis=1, inplace=True)
df_test.drop(['Item_Identifier','Outlet_Identifier'], axis=1, inplace=True)

#Print  updated Dataset
print(df_train)
print(df_test)
```

#Splitting our data into train and test

```python
from sklearn.model_selection import train_test_split

# Split the DataFrame into input and output variables
X = df_train.drop('item_outlet_sales', axis=1)
Y = df_train['item_outlet_sales']

# Split the data into training and testing sets using the train_test_split function
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=101, test_size=0.2)

# Print the training data
print(X_train)
# Print a separator
print("Test")

# Print the testing data
print(X_test)
```

#Hyper Parameter Tuning

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

# define models and parameters
model = RandomForestRegressor()
n_estimators = [10, 100, 1000]#no of trees in the model
max_depth=range(1,31)
min_samples_leaf=np.linspace(0.1, 1.0)

max_features =["auto", "sqrt", "log2"]
min_samples_split=np.linspace(0.1, 1.0, 10)


#Define the grid search object with the model, hyperparameters, and scoring metric
grid = dict(n_estimators = n_estimators)
#cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=101)


grid_search_forest = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, 
                           scoring='r2',error_score=0,verbose=2,cv=2)


grid_search_forest.fit(X_train_std, Y_train)


# summarize results
print(f"Best: {grid_search_forest.best_score_:.3f} using {grid_search_forest.best_params_}")
means = grid_search_forest.cv_results_['mean_test_score']
stds = grid_search_forest.cv_results_['std_test_score']
params = grid_search_forest.cv_results_['params']

# The 'mean_test_score' array contains the mean R-squared score obtained by cross-validation for each hyperparameter combination
#'The 'std_test_score' array contains the corresponding standard deviations. 
# The 'params' array contains a dictionary of the hyperparameters used for each combination.

for mean, stdev, param in zip(means, stds, params):
    print(f"{mean:.3f} ({stdev:.3f}) with: {param}")
# zip is a built-in function in Python that takes two or more sequences (e.g., lists, tuples, etc.) 
# and aggregates their elements into a single iterable of tuples. 
```
